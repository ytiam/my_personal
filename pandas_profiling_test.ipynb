```python
import pandas as pd
import pandas_profiling
```

### Sample Weather Data with 20000 records

```python
df = pd.read_csv('~/WA_combined.csv',nrows=20000)
```

#### Do profiling (Produce the basic report)

```python
pandas_profiling.ProfileReport(df)
```

* Click on 'Toogle Details' for each variable, to get other variable specific information.

### We can also export the profile report a XML/HTML file.

```python
profile = pandas_profiling.ProfileReport(df)
profile.to_file(outputfile="myoutputfile.html")
```
